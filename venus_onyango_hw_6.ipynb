### Question 2

#### a. Two main parts of a dynamical system
1. **State Space:** All possible values of the system's state variables, representing the various conditions the system can be in at any given time.
2. **Dynamics:** Rule or function that describes how the system transitions from one state to the next over time, allowing future states to be determined from an initial state.


#### b. Characteristics of Dynamical Systems 
1. **Stationary vs. Non-Stationary**
   - **Stationary:**  has a stable state where, given two states at different times, we cannot differentiate if one state occurred before the other (i.e., the system's state appears the same over time).
   - **Non-Stationary:** A non-stationary system has an evolving state that changes over time. In these systems, it’s possible to differentiate between states at different times. Non-stationary systems exhibit a continuous, time-dependent change in state variables.

2. **Deterministic vs. Stochastic**
   - **Deterministic:** In a deterministic system, the same initial conditions always lead to the same outcome, making future states predictable. Each state transition follows a set rule, so the system’s evolution can be precisely calculated (e.g., planetary motion).
   - **Stochastic:** A stochastic system includes randomness in its behavior, where the same initial conditions might lead to different outcomes each time. These systems are unpredictable, as random effects influence state changes, like in stock market fluctuations.

## Question 3a

```python
P0 = 110000        # Initial population
k = 0.02           # Growth rate 
K = 500000         # Carrying capacity
t_max = 100        # Maximum time
dt = 0.1           # Time step
t_values = np.arange(0, t_max + dt, dt)
P_values = np.zeros_like(t_values)
P_values[0] = P0  

for i in range(1, len(t_values)):
    P = P_values[i - 1]
    dP_dt = k * P * (1 - P / K)
    P_values[i] = P + dt * dP_dt
```

## Question 3b

```python
plt.figure(figsize=(10, 6))
plt.plot(t_values, P_values, label="Numerical Solution (Euler's Method)", )
plt.title("Population Growth Over Time Using Euler's Method")
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.legend()
plt.grid()
plt.show()
```

## Question 3c

```python

P_analytical = K / (1 + ((K - P0) / P0) * np.exp(-k * t_values))

plt.figure(figsize=(10, 6))
plt.plot(t_values, P_values, label="Numerical Solution (Euler's Method)",)
plt.plot(t_values, P_analytical, label="Analytical Solution", linestyle="--", )
plt.title("Population Growth Over Time (Numerical vs Analytical)")
plt.xlabel("Time")
plt.ylabel("Population Size")
plt.legend()
plt.grid()
plt.show()
```
